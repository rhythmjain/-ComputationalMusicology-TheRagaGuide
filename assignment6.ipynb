```python
#Change kernel to python 

from collections import Counter
import matplotlib.pyplot as plt


def get_token_tallies(file):
    with open(file, 'r') as f:
        lines = f.readlines()
    for i in range(len(lines)):
        lines[i] = lines[i].strip()
    print("totallines: ", len(lines), "unique", len(set(lines)))
    print(set(lines))
    return lines
    
def plot_hist(file_with_lines):
    plt.rcParams["figure.figsize"] = (20,3)
    w = Counter(file_with_lines)
    top = w.most_common()
    keys = [top[i][0] for i in range(len(top))]
    values = [top[i][1] for i in range(len(top))]
    plt.bar(keys, values)
    plt.show()
```

```python
adbhut_lines = get_token_tallies('adbhut_all_lines.txt')
shringar_lines = get_token_tallies('shringar_all_lines.txt')

```

```python
plot_hist(adbhut_lines)
plot_hist(shringar_lines)
```
