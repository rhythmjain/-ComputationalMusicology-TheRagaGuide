```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
!pip3 install matplotlib_venn
pip.main(["install", "matplotlib_venn"])
# Use the venn2 function
```

```python
with open('rasalist.txt', 'r') as f:
    lines = f.readlines()
    
lines = [line.rstrip() for line in lines]
```

```python
cnt = Counter()
for line in lines:
    cnt[line] += 1
tuples=cnt.most_common(10)
df = pd.DataFrame(tuples, columns = ['rasa', 'frequency'])
df

df['rasa'] = ['Shringār (Love, Erotic)', 'Bhakti (Devotion)', 'Karunā (Sad, Gloom)', 'Virah (Sad, separation)', 'Veer (Heroic)',
              'Shānt (Peace)', 'Adbhut (Wonderment)', 'Raudra (Anger)', 'Gambhir (Serious)', 'Hasya (Comic)']
```

```python
list(df[["frequency"]])

df.frequency.values.tolist(), np.sum(df.frequency.values.tolist())
```

```python

fig1, ax1 = plt.subplots()
theme = plt.get_cmap('jet')
ax1.pie(df.frequency.values.tolist(), labels=df.rasa.values.tolist(), rotatelabels=True, autopct='%1.f%%',
        shadow=True, startangle=90, textprops={"fontsize":15})
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
