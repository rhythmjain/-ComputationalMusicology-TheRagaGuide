```python
import os
import re
```

```python
def get_cleaned_lines(lines):
    final_list=[]
    for line in lines:
        if (re.search(r'[\^*]',line)):
            final_list.append(line)
            continue
        if len(line) == 1 :
            final_list.append(line)
        elif len(line) == 2:
            if line[0] == '(':
                final_list.append(line[0])
                final_list.append(line[1])
            else:
                final_list.append(line)            
        else:
            s=1
            while(s < len(line)):
                bigram = line[s-1:s+1]
                if bigram in swara_vocab:
                    final_list.append(bigram)
                    s=s+2
                else:
                    final_list.append(line[s-1])
                    s=s+1
                if s == len(line) :
                    if line[s-1] in swara_vocab:
                        final_list.append(line[s-1])
                        s= s+1
    return final_list
```

```python
path_src = 'raga_files'
path_dst = 'clean_files'
path_dst2 = 'clean_files_final'

swara_vocab = [ 'S', 'R', 'G', 'M', 'P', 'D', 'N',
                's','r','g','m','p','d','n', 
                'ss','rr','gg','mm','pp','dd','nn', 'hr','hN' ',', ')', '('
              ]

files = os.listdir(path_src)
```

```python
for file in files:
    print(f"Cleaning#1:removing unwanted characters {file}")
    str_ = f'grep -v "linebreak" {path_src}/{file} | grep -v "discrep" |sed  \'s;[q.\\\/—~];;g\' > {path_dst}/{file}'
    os.system(str_)
    


for ind in range(len(files)):
    lines = []
    print(f"Cleaning#2{files[ind]}")
    with open(f'{path_dst}/{files[ind]}') as f:
        lines=(f.readlines())
    lines = [line.strip() for line in lines]
    final_list=get_cleaned_lines(lines)
    with open(path_dst2+'/'+files[ind], 'w+') as f:
        for final_list_line in final_list:
            f.write(f'{final_list_line}\n')
    
```
