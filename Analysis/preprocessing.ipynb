```python
swara_vocab = [ 'S', 'R', 'G', 'M', 'P', 'D', 'N',
                's','r','g','m','p','d','n', 
                'ss','rr','gg','mm','pp','dd','nn', 'hr','hN' ',', ')', '('
              ]
```

```python
def get_cleaned_lines(lines):
    final_list=[]
    for line in lines:
#         print("line:", line)
        if (re.search(r'[\^*]',line)):
#             print("     Appending:", line)
            final_list.append(line)
            continue
        if len(line) == 1 :
#             print("     Appending:", line)
            final_list.append(line)
        elif len(line) == 2:
            if line[0] == '(':
                final_list.append(line[0])
                final_list.append(line[1])
            else:
                final_list.append(line)            
        else:
            s=1
            while(s < len(line)):
                bigram = line[s-1:s+1]
#                 print("     s:", s, "bigram:",bigram)
                if bigram in swara_vocab:
#                     print("     Appending:", bigram)
                    final_list.append(bigram)
                    s=s+2
                else:
#                     print("     else Appending:", line[s-1])
                    final_list.append(line[s-1])
                    s=s+1
                if s == len(line) :
#                     print("     s:", s, "line[s-1]:", line[s-1])
                    if line[s-1] in swara_vocab:
                        final_list.append(line[s-1])
#                         print("     last Appending:", line[s-1])
                        s= s+1
    return final_list
```
